```python
import matplotlib.pyplot as plt
```

```python
# Task 1: Line Plot
# Create a line plot showing the trend of monthly expenses for 6 months: [200, 250, 270, 300, 320, 310].
```

```python
expenses = [200, 250, 270, 300, 320, 310]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]

plt.plot(months, expenses, marker="o")
plt.title("Monthly Expenses")
plt.xlabel("Months")
plt.ylabel("Expenses ($)")
plt.show()
```
