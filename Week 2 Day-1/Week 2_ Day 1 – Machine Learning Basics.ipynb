```python
# Linear Regression Intuition
# Understand how linear regression predicts values:
# Plot a simple line y=2x+3 using matplotlib.
# Observe how changes in x affect y.

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = 2 * x + 3

plt.plot(x,y, label='y = 2x + 3')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Line')
plt.legend()
plt.show()
```
